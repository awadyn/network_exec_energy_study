## Wide dynamic power range
"
Current desktop and server processors can consume
**less than one-third of their peak power** at very-low activity modes, creating a **dynamic range of more than 70
percent of peak power**.
CPUs targeted at the mobile or
embedded markets can do even better, with idle power
often reaching one-tenth or less of peak power.
**In our experience, the dynamic power range of all
other components is much narrower: less than 50 percent for DRAM, 25 percent for disk drives, and 15 percent for networking switches.**
"

## Active low-power modes
"
A processor running at a lower voltage-frequency mode
can still execute instructions without requiring a performance-impacting mode transition. It is still active.
**There are no other components in the system with active low-power modes.**
Networking equipment rarely offers any
low-power modes, and the only low-power modes currently available in mainstream DRAM and disks are fully
inactive.

"

## Bad Utilization Triggers Bad Energy Efficiency

$Energy=\int\frac{Power_{instantaneous}}{dt}$

**Power** drawn by a computing component (e.g. a core) can be reduced by
1. shutting of the component entirely,
2. decreasing the frequency (and so voltage) of that component, or
3. limiting the power supply to the processor package that contains this computational component.

### C-states
A core/CPU can be incrementally idled, from an active state called **C0** where the core/CPU is executing instructions to an idle state **CX** where the core/CPU is halted or sleeping.

Other computational components on the processor chip can be idled similarly, placing the entire execution system within a range of idle behavior.



### P-States

Increasing core/CPU clock frequency should naturally increase core/CPU performance, and decreasing it should decrease performance.

**However, other factors affect this seemingly proportional relationship, like cache size and speed, memory size and speed, and interrupts. These factors could trigger bottlenecks and reduce the effects of scaling core/CPU frequency up and down.**

That being said, frequency scaling is a strongly adopted technique for power management. This is because power, frequency, and voltage are related via the following proportionality relationship:

$Power_{dynamic}  V^{2}f$

### Power Limiting
Naturally, limiting the power supply to a processor package will decrease the energy consumption of that package. **However**, this can be risky if:
1. a core/CPU on this processor package needs more power to complete its work without jeopardizing performance or
2. co-located applications on different cores/CPUs of this processor package have different power requirements, in which case *one may starve/throttle the other*, jeopardizing the performance of one.

**Frequency scaling used in tandem with power limiting can help resolve these risks.**

### Power Management via P-States
If there are potential energy savings to be had, as is the case in underutilized datacenter computational units, then the processing frequency can be reduced (and so the voltage and power drawn) to the right amount such that the computational units are better utilized (i.e. working slower but spending less idle time) and performance is not jeopardized.

**Note** that *performance is defined by the metrics that are relevant to the executing workload*. For example, performance can be measured as
1. throughput (i.e. the number of processed workload requests per unit time) or
2. latency (i.e. the time taken to process workload requests)

and the choice of which depends on the criteria of the workload being run.
